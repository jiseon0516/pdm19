### plot( )로 그래프를 그린다
> 그래프를 그리기 위해 arange 함수를 사용한다.

```python
np.arange(0.0, 2.0, 0.01) #0.0부터 2.0 미만의 수를 0.01 단위로 출력 (실수로 구성된 200개의 벡터)
len(np.arange(0.0, 2.0, 0.01))
```

```python
np.arange(0.0, 2.0, 0.01).shape #200개의 요소로 구성된 1차원 벡터, 1차원 기본 벡터의 shape은 튜플로 (200,)
```

```python
# sin그래프를 그린다
s = np.sin(np.pi*np.arange(0.0, 2.0, 0.01)) #np.pi*np.arange(0.0, 2.0, 0.01): 0~2pi까지 200등분 된 수로 구성
plt.plot(s)    # sin의 그래프를 그림
#x축을 지정하지 않고 값만 지정했기 때문에 x축의 인덱스에 문제 있음
```

### x-축을 지정하여 그리기.

```python
s1 = np.sin(np.pi*np.arange(0.0, 2.0, 0.1)) #20개의 data point
plt.plot(np.arange(0.0, 2.0, 0.1), s1, '*', ms = 8) #x축과 y축 지정// ms: marker size
#plt.plot(np.arange(0.0, 2.0, 0.01), s1, 'o'): 원이 겹쳐서 실선처럼 보임
```

## 산포도 (scatter plot)

##### sin함수는 x축이 데이터가 나온 범위, y축이 데이터 값 
##### 산포도는 x, y축  모두 데이터 값

```python
# 산포도(scatter plot)를 그린다
# 마커를 사용하여 산포도를 작성
# 표준 정규 분포에 따라 난수 생성
x = np.random.randn(5000) #데이터: 가우시안 잡음 5000개
y = np.random.randn(5000) #데이터: 가우시안 잡음 5000개
plt.figure(figsize=(5,5)) #figsize: 그림 크기 지정
t = plt.plot(x, y, 'o', alpha=0.1)   #, ms=8) #alpha: 불투명도 -> 0.1: 약간 투명하게
#가우시안 분포는 (0,0) 지점에서 가장 많이 발생
```

### [도전] 균일분포(백색잡음)의 산포도 그리기

```python
# 균일 분포에 따라 난수 생성하고 산포도 그리기
x = np.random.rand(5000) #정규잡음: randn 함수, 균일 잡음: rand함수
y = np.random.rand(5000) #데이터: 가우시안 잡음 5000개
plt.figure(figsize=(5,5)) #figsize: 그림 크기 지정
plt.plot(x, y, 'o', alpha=0.9)   #, ms=8) 
#균일분포(백색잡음)은 0~1 사이 균일하게 채워짐
```

### 막대 그래프 (Bar graph)

```python
# 막대그래프(bar graph)를 그린다.
means = [1, 2, 3] #평균
stddevs = [0.2, 0.4, 0.5] #표준편차
#평균이 1인 데이터는 표준편차 0.2, ...
bar_labels = ['bar 1', 'bar 2', 'bar 3']
#첫 번째 데이터의 레이블은 bar 1, ... => 3개의 bar

# plot bars
x_pos = list(range(len(bar_labels))) #bar 레이블의 길이만큼 => 3개이므로 0,1,2 위치
plt.bar(x_pos, means, yerr=stddevs, color=('r','g','b')) #x축은 레이블 위치, y값은 평균값, yerr는 선택(표준편차)

plt.show()
#첫 번째 bar의 중심은 0, 두 번째 bar의 중심은 1, 세 번째 bar의 중심은 2
#높이는 각각 1,2,3
#표준편차는 검은 선으로 표시됨
```

### 히스토그램 (histogram)

#### bins를 임의로 지정

```python
# 히스토그램 (histogram)
#RandomState를 이용해 seed 고정 -> 무작위 수 
#rng: random number generate
rng = np.random.RandomState(123)  # different from random.seed()
x = rng.normal(0, 20, 1000) #무작위 수 정규분표 -> 평균 0, 표준편차 20, 1000개, numpy array로

# fixed bin size
bins = np.arange(-100, 100, 5) # fixed bin size // 가우시안 분포 // 막대 개수 40개

plt.hist(x, bins=bins) #1000개의 데이터, bins=bins: 고정 단위, bins 지정 - bins = np.arange(-100, 100, 5): -100~100까지 5간격으로
plt.show()
```

### [도전하기] 위의 histogram을 정수 bins를 지정해서 그리시오.

#### bins를 정수로 지정

```python
# 히스토그램 (histogram)
rng = np.random.RandomState(123)  # different from random.seed()
x = rng.normal(0, 20, 1000) 

# fixed bin size
# bins = np.arange(-100, 100, 5) # fixed bin size
bins=100 #가우시안 막대의 개수 100개 -> 촘촘하게 그려짐, 종형 분포 유지(가우시안)

plt.hist(x, bins=bins)
plt.show()
#bins를 이용해 히스토그램 모양(구조) 변경
```

```python
# 중첩 히스토그램
rng = np.random.RandomState(123)
x1 = rng.normal(0, 20, 1000) #0 부근 중심, 벡터
x2 = rng.normal(20, 10, 1000) #20 부근 중심, 벡터
#표준편차가 작다 => 평균 값을 중심으로 더 가파르게 형성됨
```

```python
x1.shape, x2.shape 
```

```python
# fixed bin size
bins = np.arange(-100, 100, 5) # fixed bin size

plt.hist(x1, bins=bins, alpha=0.5) #파란색
plt.hist(x2, bins=bins, alpha=0.5) #빨간색(투명도 - 주황색) -> 20 부근에서 더 가파르게 분포
plt.show()
#색은 자동으로 정해서 표현
```
