## Python module 2. **matplotlib**
### Python graphics using matplotlib
- Style
- legend: 2개 이상의 그래프를 한 번에 보여줄 때 두 개의 그래프를 구분해줌
- subplot: 2개 이상의 그래프를 동시에 그림

```python
import numpy as np
import matplotlib.pyplot as plt
# 노트북 셀 내에 그림 출력
%matplotlib inline  
```

## 그림 스타일

### 그림의 스타일을 변경

```python
# 선 스타일 변경
s = np.sin(np.pi*np.arange(0.0, 2.0, 0.01))
t = plt.plot(s, linestyle=':', linewidth=4) #  test using -, --, -., :
#linestyle: 라인 모양, linewidth: 라인 두께
```

#### 선 스타일
- [line_styles_reference](https://matplotlib.org/gallery/lines_bars_and_markers/line_styles_reference.html)


### legend 표시

```python
x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x), label=('sin(x)'))
plt.plot(x, np.cos(x), label=('cos(x)'))

plt.ylabel('f(x)')
plt.xlabel('x')

#legend: 하나의 좌표에서 두 개의 그래프 그려질 때 그래프 구분하는 표식
plt.legend(loc='center')  # 'best', 'center', 'lower left'
plt.show()
```

### legend를 그림 외부에 배치하기

```python
x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x), label=('sin(x)'))
plt.plot(x, np.cos(x), label=('cos(x)'))

plt.ylabel('f(x)')
plt.xlabel('x')

plt.legend(bbox_to_anchor =(1.25, 1.))
plt.show()
```

### Multiple plot
- subplot
- subplots

```python
# import numpy as np
# import matplotlib.pyplot as plt
# plt.subplot -> single axis

def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

def g(t):
    return np.sin(np.pi*t)

t1 = np.arange(0.0, 5.0, 0.01)
t2 = np.arange(0.0, 5.0, 0.01)

plt.subplot(221)
plt.plot(t1, f(t1))

plt.subplot(222)
plt.plot(t2, g(t2))

plt.subplot(223)
plt.plot(t1, f(t1), 'r-')

plt.subplot(224)
plt.plot(t2, g(t2), 'r-')

plt.show()
```

```python
plt.subplot(211)
plt.plot(t1, f(t1))

plt.subplot(212)
plt.plot(t2, g(t2))

plt.show()
```

```python
#  multiple subplots in grids
# figure object, axis array object
x = range(11)
y = range(11)  #(11,0,-1)

fig, ax = plt.subplots(nrows=2, ncols=3,
                       sharex=True, sharey=True)  # change sharex, sharey option

for row in ax:
    for col in row:
        col.plot(x, y)
        
plt.show()
```
